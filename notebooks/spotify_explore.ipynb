```python
import pandas as pd
import glob
import json
from pathlib import Path

spotify_dir = Path("../data/spotify")
json_files = sorted(spotify_dir.glob("Streaming_History_*.json"))

dfs = []
for f in json_files:
    with open(f) as fh:
        data = json.load(fh)
    df = pd.DataFrame(data)
    df["source_file"] = f.name
    dfs.append(df)

df = pd.concat(dfs, ignore_index=True)
df["ts"] = pd.to_datetime(df["ts"])

print(f"{len(json_files)} fichiers chargés")
print(f"{len(df)} lignes au total")
print(f"Période : {df['ts'].min()} → {df['ts'].max()}")
print(f"\nColonnes ({len(df.columns)}) :")
print(df.columns.tolist())
df.head()
```
